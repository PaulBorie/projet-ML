## 1) Naive Bayes

```python
from keras.datasets import mnist
import naivebayes
import matplotlib.pyplot as plt
```

**On load le dataset MNIST qui comporte 70000 images au format 28*28 pixels.
On divise le dataset avec 85 % des images pour les données d'entrainements et 15 pourcent des images pour les données de test**

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(X_train.shape)
print(X_test.shape)
```

**on applatit chaque matrices de pixel du dataset sur une ligne ainsi chaque ligne de X_train correspond à une image du dataset (28*28 = 784 pixels), les 784 colonnes corresponent aux pixels et donc aux features dans l'algorithme naive bayes**


```python
x_train= X_train.reshape(60000, 784)
x_test=X_test.reshape(10000,784)
print(x_train.shape)
print(x_test.shape)
```

**On instancie naive Bayes avec un noise de 1000 pour la variance, on ajoute un noise à chaque variance calculée pour éviter les variances = 0 qui empêche la formule de Bayes d'être calculée car division par 0**

```python
nb = naivebayes.Bayes(1000)
nb.fit(x_train, y_train)
```

**On test notre modèle entrainée sur les 10000 images réservées au test**

```python
x_test_small = x_test[0:10000, :]
y_test_small = y_test[0:10000]
predictions, accuracy = nb.predict(x_test_small, y_test_small)
```

**On affiche 10 prédictions sur les images de test**

```python
# On affiche 10 prédictions sur les images de test
fig, axes = plt.subplots(ncols=10, sharex=False,
			 sharey=True, figsize=(20, 4))
for i in range(10):
	axes[i].set_title(predictions[i])
	axes[i].imshow(X_test[i], cmap='gray')
	axes[i].get_xaxis().set_visible(False)
	axes[i].get_yaxis().set_visible(False)
plt.show()
```

**On obtient un score de 0.815, soit 80 pr cent des images ont été bien prédites** 

```python
nb.score()
```

**Test pour trouver le noise optimal pour la variance** 

```python
x_test_small = x_test[0:300, :]
y_test_small = y_test[0:300]

scores = []
start = 950
end = 1100
for i in range (start, end):
    nb = naivebayes.Bayes(i)
    nb.fit(x_train, y_train)
    nb.predict(x_test_small, y_test_small)
    scores.append(nb.score())
max_score = max(scores)
optimal_noise_value = scores.index(max_score)
print("optimal noise value: {}".format(start + optimal_noise_value))    
```
